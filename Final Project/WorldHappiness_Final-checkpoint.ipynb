```python
#creating an if statement that values the happiness score as happy, ok, not happy.

conditions = [
    (YR2015['Happiness_Score_2015'] > 6),
    (YR2015['Happiness_Score_2015'] <= 6) & (YR2015['Happiness_Score_2015'] >0),
    (YR2015['Happiness_Score_2015'] < 4)]
choices = ['1','0','-1']
YR2015['Happiness_Bar_2015'] = np.select(conditions, choices, default='null')
YR2015
```
